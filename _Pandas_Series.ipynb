# 🐼 Pandas: A Comprehensive Introduction

## What is Pandas?

**Pandas** is an open-source Python library designed for **data manipulation, analysis, and cleaning**. It provides powerful, flexible, and easy-to-use data structures that enable users to work efficiently with structured and semi-structured data.

Built on top of **NumPy**, Pandas offers high-performance data operations while providing intuitive tools for organizing, transforming, filtering, aggregating, and visualizing data. It has become one of the most essential libraries in the Python ecosystem for data analysis and is widely used by data scientists, analysts, machine learning engineers, researchers, and software developers.

Whether working with small datasets or millions of records, Pandas simplifies complex data-processing tasks into just a few lines of code.



---

# Origin of Pandas

Pandas was created by **Wes McKinney** in **2008** while he was working at **AQR Capital Management**. During his work with financial and time-series data, he realized that Python lacked a powerful library capable of efficiently handling structured datasets similar to spreadsheets or SQL tables.

To solve this problem, he developed Pandas as an open-source project. The library quickly gained popularity because of its simplicity, flexibility, and performance. Today, Pandas is maintained by an active community of developers and contributors from around the world and has become one of the most widely used Python libraries for data analysis.

The name **"Pandas"** is derived from the term **Panel Data**, an econometrics term referring to multidimensional structured datasets.

## Dictionaries In Pandas

## Characteristics of Creating a Pandas Series from a Dictionary

- **Key-Value Structure:** A dictionary stores data in the form of key-value pairs.
- **Keys Become the Index:** When converted to a Pandas Series, the dictionary keys automatically become the index labels.
- **Values Become the Data:** The dictionary values form the data of the Series.
- **Homogeneous Data:** A Series is designed to hold values of the same data type, although Pandas also supports mixed data types when necessary.
- **Unique Keys:** Dictionary keys must be unique. Duplicate keys are not allowed.
- **Automatic Indexing:** If no custom index is provided, Pandas uses the dictionary keys as the default index.
- **Supports Custom Index:** A custom index can be specified to select or reorder data from the dictionary.
- **Mutable:** The values in the Series can be modified after creation.
- **Efficient Data Access:** Data can be accessed quickly using the corresponding index labels (dictionary keys).
- **Missing Values:** If a custom index includes keys that are not present in the dictionary, Pandas assigns `NaN` to those entries.

````python
dict1 = {'Math':98,'Science':87,'English':91,'SST':89}
marks = pd.Series(dict1)
marks
````

## Series Attribute: `dtype`

The **`dtype`** attribute returns the data type of the elements stored in a Pandas Series. It helps identify whether the Series contains integers, floats, strings, booleans, or other data types.

> **Note:** If a Series contains mixed data types, the `dtype` is usually `object`.

````python
marks = pd.Series([85, 92, 78, 90])

print(marks.dtype)
````

## Series Attribute: `shape`

The **`shape`** attribute returns the dimensions of a Pandas Series as a tuple. Since a Series is one-dimensional, it returns only the total number of elements.

> **Note:** The output is always in the form `(number_of_elements,)`.

````python
marks = pd.Series([85, 92, 78, 90])

print(marks.shape)
````

## Series Attribute: `size`

The **`size`** attribute returns the total number of elements present in a Pandas Series.

> **Note:** Unlike `shape`, it returns a single integer instead of a tuple.

````python
marks = pd.Series([85, 92, 78, 90])

print(marks.size)
````

## Series Attribute: `ndim`

The **`ndim`** attribute returns the number of dimensions of a Pandas Series. Since a Series is a one-dimensional data structure, it always returns `1`.

> **Note:** This attribute is useful when distinguishing between a Series (1D) and a DataFrame (2D).

````python
marks = pd.Series([85, 92, 78, 90])

print(marks.ndim)
````

## Series Attribute: `name`

The **`name`** attribute returns the name assigned to a Pandas Series. Naming a Series makes it easier to identify, especially when working with multiple Series or converting them into a DataFrame.

> **Note:** If no name is assigned, the `name` attribute returns `None`.

````python
marks = pd.Series([85, 92, 78, 90], name="Student Marks")

print(marks.name)
````

## Series Attribute: `empty`

The **`empty`** attribute checks whether a Pandas Series contains any elements. It returns `True` if the Series is empty; otherwise, it returns `False`.

> **Note:** A Series with `NaN` values is **not** considered empty as long as it contains elements.

````python
empty_series = pd.Series(dtype="int64")

print(empty_series.empty)
````

## Series Attribute: `hasnans`

The **`hasnans`** attribute checks whether a Pandas Series contains any missing (`NaN`) values. It returns `True` if at least one missing value exists; otherwise, it returns `False`.

> **Note:** This attribute is useful for quickly identifying missing data.

````python
marks = pd.Series([85,np.nan,78, 90])

print(marks.hasnans)
````

## Series Attribute: `axes`

The **`axes`** attribute returns a list containing the index axis of a Pandas Series. Since a Series is one-dimensional, it has only one axis.

> **Note:** The returned list contains a single `Index` object.

````python
marks = pd.Series([85, 92, 78], index=["A", "B", "C"])

print(marks.axes)
````

## Series Attribute: `T`

The **`T`** attribute returns the transpose of a Pandas Series. Since a Series is one-dimensional, transposing it does not change its structure.

> **Note:** For a Series, `T` returns the original Series itself.

````python
marks = pd.Series([85, 92, 78])

print(marks.T)
````

## Series Attribute: `array`

The **`array`** attribute returns the underlying Pandas array that stores the data in a Series. Unlike `values`, it returns a Pandas extension array instead of a NumPy array.

> **Note:** This attribute is mainly used for advanced Pandas operations.

````python
marks = pd.Series([85, 92, 78])

print(marks.array)
````

## Series Method: `memory_usage()`

The **`memory_usage()`** method returns the amount of memory consumed by a Pandas Series in bytes. It helps estimate the memory footprint of the Series.

> **Note:** By default, the returned value includes the memory used by the index.

````python
marks = pd.Series([85, 92, 78, 90])

print(marks.memory_usage())
````

## Accessing Elements with `loc[]`
 
The **`loc[]`** accessor is used to access one or more elements in a Pandas Series using **index labels** instead of integer positions. It is especially useful when the Series has custom indexes.

> **Note:** `loc[]` always works with **labels (index names)**, not numerical positions.

````python
students = pd.Series([85, 92, 78, 90], index=["Amrit", "Tahaan", "Shreyansh", "Tanmay"])
print(students.loc["Shreyansh"])
````

## Accessing Elements with `iloc[]`

The **`iloc[]`** accessor is used to access one or more elements in a Pandas Series using **integer positions**. It is useful when the exact position of an element is known.

> **Note:** `iloc[]` always works with **integer positions**, not index labels.

````python
#iloc 
print(students.iloc[3]) #Tanmay --> 90 
#Index 
#Amrit --> 0 , Tahaan --> 1, Shreyansh --> 2 , Tanmay --> 3 
````

## Slicing a Series

**Slicing** is used to retrieve a range of elements from a Pandas Series. It can be performed using index positions or index labels, depending on the method used.

> **Note:** The stop index is excluded when slicing with integer positions.

````python
numbers = pd.Series([10, 20, 30, 40, 50, 60])

print(numbers[1:5]) #Stop index is excluded
````

## Filtering a Series

Filtering is used to select elements that satisfy a specific condition. It returns a new Series containing only the values that meet the given criteria.

> **Note:** Filtering is also known as **Boolean Indexing**.

````python
#Filtering also known as Boolean Indexing 
#Used to select elements that satisfy a specific condition 
marks = pd.Series([45, 78, 92, 35, 88, 67])

print(marks[marks >= 70]) #only print marks more than or equal to  70
````

## Sorting a Series

Sorting arranges the elements of a Pandas Series either by **values** or by **index labels**. It helps organize data for easier analysis.

> **Note:** Use `sort_values()` to sort by values and `sort_index()` to sort by index labels.

````python
marks = pd.Series([78, 45, 92, 67], index=["B", "D", "A", "C"])
print(marks.sort_values()) #Ascending order 
print('-'*15)
print(marks.sort_values(ascending=False)) #Descending order 
````

## Mathematical Methods

Pandas provides several built-in mathematical methods to perform statistical and numerical calculations on a Series. These methods make it easy to summarize, analyze, and understand the data without writing complex code.

### Common Mathematical Methods

- **`sum()`** – Returns the sum of all values.
- **`mean()`** – Returns the average of the values.
- **`median()`** – Returns the middle value of the sorted data.
- **`mode()`** – Returns the most frequently occurring value(s).
- **`min()`** – Returns the smallest value.
- **`max()`** – Returns the largest value.
- **`std()`** – Returns the standard deviation of the values.
- **`var()`** – Returns the variance of the values.

> **Note:** These methods ignore missing (`NaN`) values by default.

````python
marks = pd.Series([85, 92, 78, 90, 88])

print("Sum:", marks.sum())
print("Mean:", marks.mean())
print("Median:", marks.median())
print("Mode:", marks.mode().tolist())
print("Minimum:", marks.min())
print("Maximum:", marks.max())
print("Standard Deviation:", marks.std())
print("Variance:", marks.var())
````

## String Methods

The **`str`** accessor provides access to a wide range of string methods for a Pandas Series. It allows string manipulation such as changing case, finding text, replacing characters, and calculating string lengths.

### Common String Methods

- **`str.upper()`** – Converts all strings to uppercase.
- **`str.lower()`** – Converts all strings to lowercase.
- **`str.title()`** – Converts the first letter of each word to uppercase.
- **`str.len()`** – Returns the length of each string.
- **`str.contains()`** – Checks whether a string contains a specified pattern.
- **`str.replace()`** – Replaces occurrences of a specified value with another value.

> **Note:** String methods can only be applied to Series containing string values.

````python
names = pd.Series(["shreyansh", "amrit", "tahaan", "ayush"])

print("Upper Case:")
print(names.str.upper())

print("\nTitle Case:")
print(names.str.title())

print("\nLength of Each String:")
print(names.str.len())

print("\nContains 'a':")
print(names.str.contains("a"))

print("\nReplace 'a' with '@':")
print(names.str.replace("a", "@"))
````

## Type Conversion with `astype()`

The **`astype()`** method is used to convert the data type of the elements in a Pandas Series. It is commonly used to change values from one data type to another, such as integers to floats or strings.

> **Note:** The conversion is successful only if all values in the Series can be converted to the specified data type.

````python
marks = pd.Series([85, 92, 78, 90])

print("Original Data Type:")
print(marks.dtype)

marks = marks.astype(float)

print("\nConverted Data Type:")
print(marks.dtype)

print("\nSeries:")
print(marks)
````

 ## Series Arithmetic Operations

Pandas supports arithmetic operations between two Series, such as addition, subtraction, multiplication, and division. These operations are performed **element-wise** based on the index labels.

### Common Arithmetic Operations

- **`+`** – Adds corresponding elements.
- **`-`** – Subtracts corresponding elements.
- **`*`** – Multiplies corresponding elements.
- **`/`** – Divides corresponding elements.

> **Note:** Pandas automatically aligns data based on index labels. If an index is present in one Series but not in the other, the result for that index will be `NaN`. 

````python
series1 = pd.Series([10, 20, 30], index=["A", "B", "C"])
series2 = pd.Series([5, 10, 15], index=["A", "B", "C"])

print("Addition:")
print(series1 + series2)

print("\nSubtraction:")
print(series1 - series2)

print("\nMultiplication:")
print(series1 * series2)

print("\nDivision:")
print(series1 / series2)
````

 ## Statistical Summary with `describe()`

The **`describe()`** method generates a statistical summary of a Pandas Series containing numerical data. It provides key descriptive statistics that help understand the distribution and characteristics of the data.

### Statistics Included

- **`count`** – Number of non-missing values.
- **`mean`** – Average of the values.
- **`std`** – Standard deviation.
- **`min`** – Smallest value.
- **`25%`** – First quartile.
- **`50%`** – Median (second quartile).
- **`75%`** – Third quartile.
- **`max`** – Largest value.

> **Note:** Missing (`NaN`) values are automatically ignored in the calculations.

````python
marks = pd.Series([85, 92, 78, 90, 88])

print(marks.describe())
````

## Finding Unique Values

Pandas provides methods to identify and count unique values in a Series. These methods are useful for understanding the diversity and frequency of data.

### Common Methods

- **`unique()`** – Returns all unique values in the Series.
- **`nunique()`** – Returns the total number of unique values.
- **`value_counts()`** – Returns the frequency of each unique value.

> **Note:** `unique()` returns a NumPy array, while `nunique()` returns a single integer.

````python
fruits = pd.Series(["Apple", "Banana", "Apple", "Orange", "Banana", "Apple"])

print("Unique Values:")
print(fruits.unique())

print("\nNumber of Unique Values:")
print(fruits.nunique())

print("\nFrequency of Each Value:")
print(fruits.value_counts())
````
